### Lendo dataset

base de dados retirado do site abaixo:

https://data.cityofchicago.org

```python
salarios = open('D:/Python-Projetos/DSA/dataset/salarios.csv', 'r')
dados = salarios.read()
```

```python
rows = dados.split('\n') #Separa as informações em linhas a cada quebra de linha
rows #Mostra todo conteúdo em uma unica linha
```

### Dividindo o arquivo em colunas

```python
full_data = list()
for row in rows:
    split_row = row.split(',')
    full_data.append(split_row)

full_data
```
