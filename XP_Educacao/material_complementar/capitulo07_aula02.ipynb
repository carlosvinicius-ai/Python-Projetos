Módulo **`math`**: *modulo com funções matemáticas para cálculos mais complexos*

```python
import math
print('Função cosseno:', math.cos(100))
```

```python
print('Função log:', math.log(10))
```

Módulo **`itertools`**: *modulo para construção de sequências elaboradas*

```python
import itertools

# combinação de 3 em 3
print(list(itertools.combinations([0, 1, 2, 3, 4], 3)))
```

```python
 # permutação de 2 em 2
print(list(itertools.permutations(['a', 'b', 'c'], 2)))
```

Módulo **`datetime`**: *modulo para manipulação de timestamps (datas, horários, deltas etc.)*

```python
from datetime import datetime, timedelta
print('Datetime atual:', datetime.now())
```

```python
print('Datetime após 7 dias:',  datetime.now() + timedelta(days=7))
```

Módulo **`random`**: *modulo para criação de números e sequências randômicas*

```python
import random
print('Numero aleatório entre 0 e 1:', random.random())
```

```python
print('Inteiro aleatório entre 50 e 100:', random.randint(50, 100))
```
