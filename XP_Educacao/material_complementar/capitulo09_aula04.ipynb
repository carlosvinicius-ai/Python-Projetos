## Aula 04 - Atribuição Condicional

Sintaxe:

```python
valor_se_verdadeiro if condicao else valor_se_falso
```

````python
nome = 'casa'
if len(nome) > 5:
    var = 100
else:
    var = 0
print('O valor de var é:', var)
````

````python
nome = 'casa'
var = 100 if len(nome) > 5 else 0
print('O valor de var é:', var) 
````
