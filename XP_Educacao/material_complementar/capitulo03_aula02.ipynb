```python
x = 10
y = 4
z = 4.5
print('x + y + z=', x + y + z)
print('x - 3 =', x - 3)
print('x * z =', x * z)
print('x * 2 =', x * 2)
print('x / (y + 2) =', x / (y + 2))
print('x % y =', x % y)
print('x // y =', x // y)
print('x ** z =', x ** z)
```

### Operadores de Comparação

```python
x = 5
y = 1
s = 'palavra'

print('x > y:', x > (y + 2))
print('x <= 4.564:', x <= (4.564 + 1))
print('s == palavra:', s != 'palavra')
print('(y * 0) == False:', (y * 0) == False)
print('s != abacaxi:', s != 'abacaxi')
```

Comparação com strings

```python
x = 3
y = 'palavra'
print(x == y)
print(x != y)
#print(x > y)
```

```python
a = 'casa'
b = 'predio'
c = 'elevador'
d = 'casamento'
print(a < b)
print(b < c)
print(b > c)
print(a > d)
print('abacaxi' >= 'abacaxi')
```
