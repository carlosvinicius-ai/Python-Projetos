## Aula 04 - Conjuntos (`set`)
- Criação de conjuntos
- Operações com conjuntos

Em Python os conjuntos (sets) são estruturas de dados do tipo set, não-ordenadas, que representam uma coleção de itens únicos, ou seja, itens sem repetições. 

Assim como a lista, o conjunto também é uma estrutura mutável, suportando operações de inserção, alteração e remoção de elementos. 

A diferença fundamental entre as listas e os conjuntos é que os elementos dos conjuntos são únicos e a sua ordem dentro da estrutura não importa. Por isto eles são ditos como estruturas não-ordenadas. 

Adicionalmente, o tipo set suporta operações matemáticas entre conjuntos como união, interseção e diferença.

### Operacões com Conjuntos


image.png

```python
# Criação dos conjuntos A e B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print('A:', A)
print('B:', B)
```

```python
# Operação de União: (A ∪ B)
print('A | B =>', A | B)
print('A.union(B) =>', A.union(B))
```

```python
# Operação de Interseção: (A ∩ B)
print('A & B =>', A & B)
print('A.intersection(B) =>', A.intersection(B))
```

```python
# Operação de Diferença: (A - B) e (B - A)
print('A - B =>', A - B)
print('A.difference(B) =>', A.difference(B))
print('B - A =>', B - A)
print('B.difference(A) =>', B.difference(A))
```

```python
len(A - B)
```

```python
# Criação dos conjuntos c1, c2 e c3
c1 = {1, 2, 3, 4, 5}
c2 = {4, 5}
c3 = {91, 92, 93}

# Adiciona um elemento ao conjunto
c1.add(6)
print(c1)

# Adiciona os elementos de uma sequência iterável
c1.update([2, 4, 6, 8])
print(c1)
```

```python
c1.remove(1)
print(c1)
```

```python
# Descarta um elemento do conjunto, 
c1.discard(8)
print(c1)

# Diferentemente do set.remove(), o discard não gera um erro 
# se o elmento a ser removido não existir
c1.discard(99)
print(c1)
```

```python
# Verifica se os conjuntos são disjuntos, ou seja,
# se não possuem nenhum elemento em comum
print(c1.isdisjoint(c2))
print(c1.isdisjoint(c3))

# Verifica se o conjunto é subconjunto de outro
print(c1.issubset(c2))
print(c2.issubset(c1))

# Verifica se o conjunto contém outro conjunto (superset)
print(c1.issuperset(c2))
print(c2.issuperset(c1))
```

### Exemplos práticos com conjuntos

Considere que estamos desenvolvendo um programa para gerenciar uma escola de línguas estrangeiras que possui três turmas de alunos para os seguintes idiomas: inglês, espanhol e francês. Considere também, que um mesmo aluno pode estar matriculado em uma única turma ou em múltiplas turmas, conforme as listagens abaixo:

image.png

Neste momento estamos interessados em resolver dois problemas desta escola:

1.   Criar uma relação com todos os alunos da escola, sem repetições;
2.   Identificar os alunos que estão matriculados em pelo menos duas turmas para oferecermos um desconto;

```python
# Resolução do problema 01 com listas
# RESULTADO TERÁ ALUNOS REPETIDOS

# Criação das listas de alunos das turmas
ING = ['Gabriel', 'Caio', 'Maria', 'Ana', 'Juliano', 'Flávia', 'Rubens', 'Bruna']
ESP = ['Caio', 'Artur', 'Beatriz', 'Carolina', 'Maria', 'Juliano', 'Bruna', 'Rui']
FRA = ['Pedro', 'Bruna', 'Paula', 'Tiago', 'Maria', 'Flávia', 'Rui', 'Carolina']

# Concatenação de todas as listas de alunos
TODOS = ING + ESP + FRA

# Ordenação para melhor visualização
TODOS.sort()

# Exibição do resultado
print('Relação de todos os alunos da escola:')
print(TODOS)
```

```python
# Resolução do problema 01 com conjuntos
# RESULTADO NÃO TERÁ ALUNOS REPETIDOS

# Criação das listas de alunos das turmas
ING = {'Gabriel', 'Caio', 'Maria', 'Ana', 'Juliano', 'Flávia', 'Rubens', 'Bruna'}
ESP = {'Caio', 'Artur', 'Beatriz', 'Carolina', 'Maria', 'Juliano', 'Bruna', 'Rui'}
FRA = {'Pedro', 'Bruna', 'Paula', 'Tiago', 'Maria', 'Flávia', 'Rui', 'Carolina'}

# União de todas os conjuntos de alunos
TODOS = ING | ESP | FRA

# # Ordenação para melhor visualização
#TODOS.sort()

# Exibição do resultado
print('Relação de todos os alunos da escola:')
print(TODOS)
```

```python
# Resolução do problema 02 com conjuntos

ING = {'Gabriel', 'Caio', 'Maria', 'Ana', 'Juliano', 'Flávia', 'Rubens', 'Bruna'}
ESP = {'Caio', 'Artur', 'Beatriz', 'Carolina', 'Maria', 'Juliano', 'Bruna', 'Rui'}
FRA = {'Pedro', 'Bruna', 'Paula', 'Tiago', 'Maria', 'Flávia', 'Rui', 'Carolina'}

# 1 – Interseção entre os pares de turmas: (ING & ESP), (ING & FRA) e (ESP & FRA)
# 2 – Calcula a união das interseções
ALUNOS_DESCONTO = (ING & ESP) | (ING & FRA) | (ESP & FRA)

# Exibição do resultado
print('Relação de dos alunos com desconto:')
print(ALUNOS_DESCONTO)
```
