## Aula 03 - Argumentos das Funções


```python
# calcula a área de um triângulo: (b x h) / 2
def area_triangulo(base, altura): 
    return (base * altura)/2

print(area_triangulo(10, 5))
```

```python
def teste(a, b):
    return a - b
```

```python
x = 10
y = 4
teste(y, x)
```

```python
teste(10, b=3)
```

```python
print(area_triangulo(altura=10, base=5))
```

```python
print(area_triangulo(5))
```

```python
#Argumento padrão (default)
def exibe_pessoa(nome, idade=30): 
    print(f'Meu nome é {nome} e tenho {idade} anos.')

exibe_pessoa('Antônio')
```

```python
exibe_pessoa('Antônio', 36)
```
