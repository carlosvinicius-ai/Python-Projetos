## Aula 05 - Dicionários (`dict`)
- Criação dos dicionários e acesso aos elementos
- Operações com dicionários
- Iteração sobre dicionários

Em Python, um dicionário, que é do tipo de dados dict, também é uma coleção de itens, entretanto cada elemento é um par key – value (chave – valor). Estes pares, indicam que cada elemento possuí um valor/value atrelado à uma chave/key.

### Criação dos Dicionários e Acesso dos Elementos

```python
#{chave1:valor1, chave2:valor2, chave3:valor3, chave4:valor4}
```

```python
# Dicionário onde as chaves são do tipo string
d1 = {'nome': 'Antônio', 'idade': 36, 'sexo': 'masculino'}
print(d1)
```

```python
# Dicionário onde as chaves são do tipo inteiro
d2 = {2: 'dois', 1: 'um', 4: 'quatro', 3: 'três', 0: 'zero'}
print(d2)
```

```python
# Dicionário com chaves e valores de tipos mistos
d3 = {2: 'a', 5.44: True, 'key': None}
print(d3)
```

```python
# Dicionário vazio
d4 = {}
print(d4)
```

```python
# Criação dos dicionários
d1 = {2: 'dois', 1: 'um', 4: 'quatro', 3: 'três', 0: 'zero', True: 'casa'}
d2 = {'nome': 'Antônio Carlos', 'idade': 36, 'sexo': 'masculino'}

# Acesso aos elementos
print(d1[0])
print(d1[2])
print(d1[True])
```

```python
print(f'Meu nome é {d2["nome"]} e tenho {d2["idade"]} anos')
```

### Operacões com Dicionários

```python
# Criação do dicionário
d1 = {'nome': 'Antônio', 'idade': 36, 'sexo': 'masculino'}
print(d1)
```

```python
# Atualização do valor da chave 'nome'
d1['nome'] = 'Antônio Carlos'
print(d1)
```

```python
# Adição da chave 'endereço'
d1['endereço'] = 'Rua 123'
print(d1)
```

```python
# Cria o dicionário
d1 = {'zero': 0, 'um': 1, 'dois': 2, 'três': 3, 'quatro': 4}
print(d1)
```

```python
# Encontra a maior e menor chave
print('Maior e menor chave:', max(d1), min(d1))
```

```python
# Adiciona elementos de um outro dicionario
d1.update({'cinco': 5, 'seis': 6})
print(d1)
```

```python
# Verifica se o dicionário possui as seguintes chaves
print("A chave 'dois' está no dicionário?", 'dois' in d1)
print("A chave 5 não está no dicionário?", 5 in d1)
```

```python
# Remove o elemento com chave 'zero'
d1.pop('seis')
print(d1)
```

### Iteração sobre os dicionários

```python
# Cria o dicionário
d1 = {'zero': 0, 'um': 1, 'dois': 2, 'três': 3, 'quatro': 4, 'dez': 10}

# Itera sobre as chaves
for key in d1:
    print(d1[key])
    if key == 'três':
        print('Chave três encontrada!')
        break

for key in d1.keys():
    if key == 'dez':
        print('Chave dez encontrada!')
        break
```

```python
for key in d1:
    print(key, d1[key])
```

```python
# Itera sobre os valores
soma = 0
for value in d1.values():
    soma = soma + value
print('Soma dos valores do dicionário:', soma)
```

```python
list(d1.values())
```

```python
# Itera sobre os itens
for key, value in d1.items():
    print(key, value)
```

```python
list(d1.items())
```
