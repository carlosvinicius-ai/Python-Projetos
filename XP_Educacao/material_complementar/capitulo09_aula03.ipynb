## Aula 03 - Funções Anônimas (Funções Lambda)

Sintaxe:

```python
lambda argumentos: <expressão>
```

````python
# Declaração da função explicita (declaração tradicional)
def area_quadrado(lado):
    return lado * lado
````

````python
area_quadrado(5)
````

````python
# Declaração da função lambda
area_quad = lambda lado: lado * lado

# Utilização
area_quad(4)
````

````python
area_triangulo = lambda base, altura: (base * altura)/2
area_triangulo(10, 20)
````

````python
#Verificação do tipo da variável
type(area_quadrado)
````

````python
# Função que calcula o triplo de número
triplo = lambda x: x * 3
triplo(10)
````

````python
# Calcula o triplo dos números de uma lista
lista = [4, 5, 9, 7, 0, 1, 8]
list(map(triplo, lista))
````

````python
# Podemos fazer de um jeito mais pythonico ainda!
list(map(lambda x: x ** 3, [4, 5, 9, 7, 0, 1, 8]))
````
